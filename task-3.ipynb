## Importing Modules

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import tree
```

## Data Extracting

```python
df = pd.read_csv("iris.csv",index_col = 'Id')
```

```python
df
```

```python
x = df.iloc[:,[0,1,2,3]]
```

```python
x
```

```python
y = df.iloc[:,[4]]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
gender_encoded = le.fit_transform(y['Species'])
print(gender_encoded)
```

```python
y = gender_encoded
```

```python
y
```

## Generating data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 3)
```

## Generating the decision tree model using Entropy as criteria

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 3)
classifier.fit(X_train, y_train)
```

## Getting confusion matrix and accuracy of model(Entropy)

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
fig = plt.figure(figsize=[10,10])
sns.heatmap(cm,annot = True)
```

## Visualising data uing tree plot(Entropy)

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier)
plt.show()
```

```python
fn=x.columns
cn=df['Species'].unique()
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(classifier,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

## Generating the decision tree model using Gini as criteria

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'gini', random_state = 3)
clf.fit(X_train, y_train)
```

## Getting confusion matrix and accuracy of model(Gini)

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred1 = clf.predict(X_test)
cm2 = confusion_matrix(y_test, y_pred1)
print(cm2)
accuracy_score(y_test, y_pred1)
fig = plt.figure(figsize=[10,10])
sns.heatmap(cm2,annot = True)
```

## Visualising data uing tree plot(Gini)

```python
fn=x.columns
cn=df['Species'].unique()
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```
